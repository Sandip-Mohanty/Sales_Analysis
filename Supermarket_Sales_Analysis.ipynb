```python
# Convert 'DATE' column to datetime
txn_ds['DATE'] = pd.to_datetime(txn_ds['DATE'], errors='coerce')

# Convert numerical columns to appropriates data types
txn_ds['TOT_SALES'] = pd.to_numeric(txn_ds['TOT_SALES'],errors='coerce')
txn_ds['PROD_QTY'] = pd.to_numeric(txn_ds['PROD_QTY'],errors='coerce')

# Convert categorical coolumns to Categories data types 
txn_ds['STORE_NBR'] = txn_ds['STORE_NBR'].astype('category')
txn_ds['LYLTY_CARD_NBR'] = txn_ds['LYLTY_CARD_NBR'].astype('category')
txn_ds['PROD_NBR'] = txn_ds['PROD_NBR'].astype('category')
txn_ds['PROD_NAME'] = txn_ds['PROD_NAME'].astype('category')

txn_ds.dtypes
```

```python
cust_ds = pd.read_csv('QVI_purchase_behaviour.csv')
cust_ds
```

```python
# Identify key column
print('Identified key columns:\n')
print('Customer Id Column:','LYLTY_CARD_NBR')
print('Marital Status:','LIFESTAGE')
print('Class Of Customer:','PREMIUM_CUSTOMER')
```

```python
# Handling missing values
cust_ds.isna().sum()
```

```python
# Check data type of customer dataset
cust_ds.dtypes
```

```python
# Data Formating
cust_ds['LYLTY_CARD_NBR']=pd.to_numeric(cust_ds['LYLTY_CARD_NBR'],errors='coerce')
cust_ds['LIFESTAGE']=cust_ds['LIFESTAGE'].astype('category')
cust_ds['PREMIUM_CUSTOMER']=cust_ds['PREMIUM_CUSTOMER'].astype('category')

cust_ds.dtypes
```

```python
# Merge transaction and customer data 
merge_df = pd.merge(cust_ds,txn_ds,on='LYLTY_CARD_NBR',how='inner')
merge_df
```

```python
merge_df.to_csv('Merge_Dataset.csv',index=False)
```

```python
# Check the structure of Merged dataset 
print('Merge dataset information:\n')
merge_df.info()
```

# Calculate Metrics

```python
# Customer segmentation and analysis

# Calculate total sales
total_sales = merge_df['TOT_SALES'].sum()
print('Total Sale:',total_sales)

```

```python
# Calculate average sales per segment
average_sales_per_segment = merge_df.groupby(['LIFESTAGE','PREMIUM_CUSTOMER'])['TOT_SALES'].mean()
average_sales_per_segment
```

```python
# Count the number of transaction per segment
transactions_per_segment=merge_df.groupby(['LIFESTAGE','PREMIUM_CUSTOMER'])['TXN_ID'].count()
transactions_per_segment
```

```python
# Merge the segment data
segment_analysis = pd.merge(transactions_per_segment, average_sales_per_segment, on=['LIFESTAGE', 'PREMIUM_CUSTOMER'])
segment_analysis.head()
```

# Visualization

```python
sales_by_lifestage = merge_df.groupby('LIFESTAGE')['TOT_SALES'].sum().reset_index()

plt.figure(figsize=(10, 6))
bp = plt.bar(sales_by_lifestage['LIFESTAGE'], sales_by_lifestage['TOT_SALES'])
for bar in bp:
    yval = bar.get_height()  # Get the height (value) of each bar
    plt.text(bar.get_x() + bar.get_width() / 2, yval, round(yval, 2), 
             ha='center', va='center_baseline',rotation=90)
plt.xlabel('LIFESTAGE')
plt.ylabel('TOTAL SALES')
plt.title('Total Sales by Lifestage')
plt.xticks(rotation=25)
plt.show()
```

```python
sales_by_premium = merge_df.groupby('PREMIUM_CUSTOMER')['TOT_SALES'].sum().reset_index()

plt.figure(figsize=(10, 6))
bars = plt.bar(sales_by_premium['PREMIUM_CUSTOMER'], sales_by_premium['TOT_SALES'],color = 'y')
for bar in bars:
    yval = bar.get_height()  # Get the height (value) of each bar
    plt.text(bar.get_x() + bar.get_width() / 2, yval, round(yval, 2), 
             ha='center', va='center_baseline')
plt.xlabel('PREMIUM CUSTOMER')
plt.ylabel('TOTAL SALES')
plt.title('Total Sales by Premium Customer')
plt.show()
```

```python
sales_by_product = merge_df.groupby('PROD_NAME')['TOT_SALES'].sum().reset_index()

plt.figure(figsize=(18,8))
bar1 = plt.bar(sales_by_product['PROD_NAME'], sales_by_product['TOT_SALES'],color = 'g')

for bar in bar1:
    yval = bar.get_height()  # Get the height (value) of each bar
    plt.text(bar.get_x() + bar.get_width() / 2, yval, round(yval, 2), 
             ha='center', va='bottom',rotation=90)  # Position the label at the center top
plt.xlabel('PRODUCT NAME')
plt.ylabel('TOTAL SALES')
plt.title('Total Sales by Product')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12,18))

product_lifestage_count = merge_df.groupby('PROD_NAME')['LIFESTAGE'].count().reset_index()
bar2 = plt.barh(product_lifestage_count['PROD_NAME'],product_lifestage_count['LIFESTAGE'], color='skyblue')

for bar in bar2:
    xval = bar.get_width()  # Get the height (value) of each bar
    plt.text(xval,bar.get_y() + bar.get_height() / 2, round(xval, 2), 
             ha='left', va='bottom')  # Position the label at the center top
plt.xlabel('PRODUCT AMOUNT')
plt.ylabel('PRODUCT NAME')
plt.title('Total Amount of Product Sale')

plt.show()
```

```python
plt.figure(figsize=(12, 8))
ax = segment_analysis.pivot_table(index="LIFESTAGE", columns="PREMIUM_CUSTOMER",values= "TXN_ID")

bar_plot = ax.plot(kind='bar', stacked=True)
# Add labels to each bar in the stacked bar plot
for container in bar_plot.containers:
    # Get the data labels for each segment (stack)
    bar_plot.bar_label(container, label_type='center')
    
plt.xlabel('LIFESTAGE')
plt.ylabel('TRANSACTION COUNT')
plt.title('Transaction Count per Customer Segment')
plt.xticks(rotation='vertical')
plt.show()
```

```python
plt.figure(figsize=(12, 18))
average_sales_per_segment = segment_analysis.pivot_table(index="LIFESTAGE",columns= "PREMIUM_CUSTOMER", values="TOT_SALES")
bar1_plot = average_sales_per_segment.plot(kind='barh', stacked=True)

for container in bar1_plot.containers:
    # Get the data labels for each segment (stack)
    bar1_plot.bar_label(container, label_type='center')
    
plt.xlabel('LIFESTAGE')
plt.ylabel('AVERAGE SALES')
plt.title('Average Sales per Customer Segment')
plt.xticks(rotation='vertical')
plt.show()
```
